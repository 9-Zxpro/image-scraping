```python
import requests
from PIL import Image
from io import BytesIO
from reportlab.lib.pagesizes import letter
from reportlab.lib.utils import ImageReader
from reportlab.pdfgen import canvas
from bs4 import BeautifulSoup
import tempfile
import os

def create_pdf_from_linkedin_carousel(html_code, output_pdf="ds_interview_book.pdf"):
    soup = BeautifulSoup(html_code, 'html.parser')
    img_tags = soup.find_all('img', class_='carousel-lazy-element')
    image_urls = [img['data-src'] for img in img_tags if 'data-src' in img.attrs]

    if not image_urls:
        print("No image URLs found in the provided HTML.")
        return

    c = canvas.Canvas(output_pdf, pagesize=letter)

    for i, img_url in enumerate(image_urls):
        temp_image_path = None
        try:
            print(f"Downloading image {i+1}/{len(image_urls)}: {img_url}")
            response = requests.get(img_url, stream=True)
            response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
            image_data = BytesIO(response.content)
            img = Image.open(image_data)
            img_width, img_height = img.size
            pdf_width, pdf_height = letter

            # Calculate aspect ratios to fit the image within the PDF page
            img_ratio = img_width / img_height
            pdf_ratio = pdf_width / pdf_height

            if img_ratio > pdf_ratio:
                # Image is wider than PDF, fit by width
                draw_width = pdf_width
                draw_height = pdf_width / img_ratio
                x_offset = 0
                y_offset = (pdf_height - draw_height) / 2
            else:
                # Image is taller than or equal aspect ratio to PDF, fit by height
                draw_height = pdf_height
                draw_width = pdf_height * img_ratio
                x_offset = (pdf_width - draw_width) / 2
                y_offset = 0

            image_reader = ImageReader(image_data)
            c.drawImage(image_reader, x_offset, y_offset, width=draw_width, height=draw_height)
            c.showPage()

        except requests.exceptions.RequestException as e:
            print(f"Error downloading image from {img_url}: {e}")
        except Exception as e:
            print(f"Error processing image {img_url}: {e}")
        finally:
            # Clean up the temporary file
            if temp_image_path and os.path.exists(temp_image_path):
                os.remove(temp_image_path)

    c.save()
```

```python
from pdf2image import convert_from_path
import pytesseract
from PyPDF2 import PdfMerger
from io import BytesIO

INPUT_PDF  = '/kaggle/input/data-sci-qa/ds_interview_book.pdf'
OUTPUT_PDF = '/kaggle/working/ds_interview_text.pdf'

def image_pdf_to_searchable(input_pdf, output_pdf, dpi=300):
    pages = convert_from_path(input_pdf, dpi=dpi)
    merger = PdfMerger()
    for i, page_img in enumerate(pages, start=1):
        pdf_bytes = pytesseract.image_to_pdf_or_hocr(page_img, extension='pdf')
        merger.append(BytesIO(pdf_bytes))
        print(f"Processed page {i}/{len(pages)}")
    with open(output_pdf, "wb") as fout:
        merger.write(fout)
    print(f"‚úÖ Saved searchable PDF to: {output_pdf}")

image_pdf_to_searchable(INPUT_PDF, OUTPUT_PDF)
```
